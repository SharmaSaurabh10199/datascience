```python
# load the data
import json 
def load_data():
    with open("massive_data.json",'r') as file:
        return json.load(file)
```

```python
dataset= load_data()
```

```python
## find the peoplw you may know 
def people_you_may_know(user_id, data):
    # find the dictionalry contaning the friends
    user_friends={}
    for user in data['users']:
        user_friends[user['id']]= set(user['friends'])
    
    direct_friends= [user['friends'] for user in data['users'] if user['id']==user_id]
    # find the friends of friends
    suggested={}
    for mutual in user_friends[user_id] :
        for possible_option in user_friends[mutual]:
            if (user_id != possible_option) and (possible_option not in user_friends[user_id]):
                suggested[possible_option]= suggested.get(possible_option,0)+1

    # the key is possible options of the friends 
    # sort this on the basis of suggested value 
    sorted_values= list(sorted(suggested.items(),key= lambda x: x[1], reverse=True))
    print(sorted_values)
```
