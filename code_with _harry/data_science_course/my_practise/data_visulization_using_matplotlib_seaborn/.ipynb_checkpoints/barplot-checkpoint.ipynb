```python
# bar charts
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.bar(years,runs)
# plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Sachins runs over the years')
# plt.legend()
```

```python
# plotting bar charts side by side 
# plot the runs of sachin, kohli,sehwag side by side on bar chart
import numpy as np
 
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

# find common index as x 
x=np.arange(len(years))
x

plt.bar(years, sachin)
plt.bar(years, sehwag)
plt.bar(years, kohli)
# the problem with this is that its hard to comapre the runs 
```

```python
# plotting in correct way 
width=0.35
plt.bar(x-width, sachin)
plt.bar(x, sehwag)
plt.bar(x+width, kohli)
```

```python
# sincethe widh of the bars is more than then the difference between the axis , 
# this things are overlapping 
plt.bar(x-width,sachin,width=width, label="Sachin" )
plt.bar(x, sehwag,width=width, label="Sehwag" )
plt.bar(x+width,kohli,width=width ,  label="Kohli")
plt.legend()
```

```python
# xticks: shows actual years '
plt.xticks(x,years)
plt.show()
```

```python
# horizontal bar charts 
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

# plotting horizontal plot 
plt.barh(players,runs_5yrs)
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```

```python
# adding valur labels with plt.text()
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players,runs)
plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Runs of different players")

# putting actual values on the top of bar charts 

for i in range(len(players)):
    plt.text(i,runs[i] + 20, str(runs[i]), ha='center')
```
