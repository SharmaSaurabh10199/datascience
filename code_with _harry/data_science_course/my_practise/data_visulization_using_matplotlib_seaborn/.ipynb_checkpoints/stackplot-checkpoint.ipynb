```python
# Stackplots in Matplotlib are a type of area chart that show how multiple data series contribute to a whole over time or another continuous variable. 

import matplotlib.pyplot as plt 

days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels=['studying','playing','watching_tv','sleeping']
plt.stackplot(days, studying,playing,watching_tv,sleeping, labels=labels)

plt.legend()
plt.legend(loc='upper left')
```
