```python
# subplots 
import matplotlib.pyplot as plt 
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]


 
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('x*2')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('x**2')
```

```python
# plots as array 
fig, axs= plt.subplots(1,2)

axs[0].plot(x,y1)
axs[0].set_title("x*2")

axs[1].plot(x,y2)
axs[1].set_title("x**2")

# title of the overall plot 
fig.suptitle('Simple Comparison Plots', fontsize=14)
```
