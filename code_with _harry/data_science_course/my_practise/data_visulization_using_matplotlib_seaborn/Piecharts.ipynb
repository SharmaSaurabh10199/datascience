```python
# these are used for lesser number of categories
import matplotlib.pyplot as plt

# basic pie chart 
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.pie( runs,labels=labels)
```

```python
# custom colors 
colors=['red','blue','green','pink']
plt.pie( runs,labels=labels, colors=colors)
```

```python
# wedgeprops: 
plt.pie(
    runs,
    labels=labels,
    wedgeprops={
        'edgecolor': 'black',
        'linewidth': 2,
        'linestyle': 'dashed',
        'antialiased': True
    }
)
```

```python
# Explode : used to take out the slice at specific 

explode=[0.1,0,0,0]
plt.pie( runs,labels=labels, explode=explode)
```

```python
# shadow adding 
plt.pie( runs,labels=labels, explode=explode, shadow=True)
```

```python
# rotate the pie chart at a certain angle 
plt.pie( runs,labels=labels, explode=explode, startangle=180)
```

```python
# atopct: to show the percentage 
plt.pie( runs,labels=labels, explode=explode, autopct='%1.1f%%')
```

```python
# - '%1.1f%%' → shows one decimal place (e.g., 25.0%)
# - '%1.0f%%' → shows whole numbers (e.g., 25%)
```
