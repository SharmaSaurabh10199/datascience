```python
import pandas as pd 
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python

departments
```

```python
employees
```

```python
# inner join 
pd.merge(employees,departments,on='DeptID')
```

```python
# left join 
pd.merge(employees,departments,on='DeptID' , how='left')
```

```python
# right join 
pd.merge(employees,departments,on='DeptID', how='right')
```

```python
# outer join : gets all the rows 
pd.merge(employees,departments,on='DeptID', how='outer')
```

```python
# concatinae: its basicaly union all 
pd.concat([departments,employees])
```

```python
# on axis=1 : i:e along the columns 
pd.concat([departments,employees], axis=1)
```
